### Exercise 4: Pie Chart with Percentages

```python
import matplotlib.pyplot as plt

# Define data
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [10, 15, 7, 5]

# Error handling and validation
try:
    # Check that sizes are all numbers
    if not all(isinstance(size, (int, float)) for size in sizes):
        raise ValueError("All sizes must be numbers (int or float).")

    # Check that sizes and labels match
    if len(labels) != len(sizes):
        raise ValueError("Labels and sizes must be the same length.")

    # Check that total > 0 to avoid zero division
    total = sum(sizes)
    if total <= 0:
        raise ValueError("Sum of sizes must be greater than zero.")

    # Normalize sizes to percentages (optional but improves clarity)
    normalized_sizes = [size / total * 100 for size in sizes]

    # Create the pie chart
    plt.figure(figsize=(6, 6))
    plt.pie(normalized_sizes, labels=labels, autopct='%1.1f%%', startangle=140)
    plt.title('Fruit Pie Chart')
    plt.axis('equal')  # Equal aspect ratio ensures pie is drawn as a circle.

    plt.show()

except Exception as e:
    print(f"Error: {e}")
```
